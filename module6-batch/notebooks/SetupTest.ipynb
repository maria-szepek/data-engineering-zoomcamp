```python
# that is a thing that tells how we connect to spark
from pyspark.sql import SparkSession
```

```python
# define how we connect to spark: we will use local master using all the available cpu
spark = SparkSession.builder \
    .master("local[*]") \
    .appName("test") \
    .getOrCreate()

```
